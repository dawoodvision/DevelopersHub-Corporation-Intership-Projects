```python
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load the iris dataset
df = sns.load_dataset("iris")
df
```

```python
df.head() # First 5 rows
```

```python
# Check the structure of the dataset
print("Shape of dataset:", df.shape)   # Number of rows and columns
print("\nColumns:", df.columns)        # Column names
```

```python
# check for missing values
print("\nMissing values:\n", df.isnull().sum())
```

```python
# dataset info
print("\nDataset Info:")    
df.info()
```

# 📊 Dataset Understanding

The **Iris dataset** has 150 rows and 5 columns: sepal length, sepal width, petal length, petal width, and species.  

**Key points:**  
- No missing values  
- All features are numeric except `species`  
- We can explore feature relationships, distributions, and patterns

```python
# 📈 Scatter Plot: Sepal Length vs Sepal Width

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species")
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

# 📊 Scatter Plot Data Insights

- **Setosa (blue):** short and wide sepals  
- **Versicolor (orange) & Virginica (green):** long and narrow sepals  

✅ Sepal dimensions clearly help **distinguish between species**

```python
# 📊 Histogram: Sepal Length Distribution
plt.figure(figsize=(8,6))
sns.histplot(data=df, x="sepal_length", bins=20, kde=True, hue="species")
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Count")
plt.show()
```

# 📊 Histogram Insights

- The histogram of **sepal length** shows a clear separation between the three iris species.  
- **Setosa** has a peak around **5.0 cm**  
- **Versicolor** peaks around **5.5 cm**  
- **Virginica** peaks around **6.5 cm**  

✅ This shows that **sepal length** is a useful feature to distinguish between species.

```python
# 📦 Box Plot of Iris Features
plt.figure(figsize=(8,6))
sns.boxplot(data=df.drop("species", axis=1), palette="pastel")
plt.title("Box Plot of Iris Features")
plt.show()
```

# 📦 Box Plot Insights

- Sepal and petal lengths have a wider range.  
- Sepal and petal widths are more clustered.  
- Some outliers appear in sepal width.  
- Box plots show spread and variation of features.

```python
# Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.drop("species", axis=1).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Iris Features")   
plt.show()
```

# 🔥 Correlation Heatmap Insights

- There is a **strong positive correlation** between **Sepal Length** and **Petal Length**.  
- This means that as sepal length increases, petal length tends to increase too.
